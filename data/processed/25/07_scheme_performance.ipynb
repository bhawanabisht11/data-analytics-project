```python
#1. Validate return columns are numeric
```

```python
return_cols = [
    'return_1yr_pct',
    'return_3yr_pct',
    'return_5yr_pct',
    'benchmark_3yr_pct'
]

for col in return_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print(df[return_cols].dtypes)
```

```python
invalid_returns = df[df[return_cols].isnull().any(axis=1)]

print("Rows with invalid return values:", len(invalid_returns))
```

```python
#2. Flag anomalous return values
```

```python
# Mutual fund returns beyond ±100% are unusual and should be reviewed.
```

```python
anomalies = df[
    (df[return_cols] > 100).any(axis=1) |
    (df[return_cols] < -100).any(axis=1)
]

print("Anomalous return rows:", len(anomalies))
display(anomalies)
```

```python
# 3. Validate expense ratio range (0.1%–2.5%)
```

```python
expense_anomalies = df[
    (df['expense_ratio_pct'] < 0.1) |
    (df['expense_ratio_pct'] > 2.5)
]

print("Expense ratio anomalies:", len(expense_anomalies))
display(expense_anomalies[
    ['amfi_code', 'scheme_name', 'expense_ratio_pct']
])
```

```python
# I am Using sqlite3
```

```python
import sqlite3
```

```python
# Create/connect database
conn = sqlite3.connect('mutual_fund.db')
# Create cursor
cursor = conn.cursor()
```

```python
# Execute SQL
cursor.execute("""
CREATE TABLE dim_fund (
    fund_key INTEGER PRIMARY KEY AUTOINCREMENT,
    amfi_code INTEGER UNIQUE NOT NULL,
    scheme_name TEXT NOT NULL,
    fund_house TEXT NOT NULL,
    category TEXT,
    plan TEXT,
    risk_grade TEXT
);
""")
```

```python
# Save changes
conn.commit()

print("Table created successfully!")

# Close connection
conn.close()
```

```python
conn = sqlite3.connect("mutual_fund.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE dim_date (
    date_key INTEGER PRIMARY KEY,
    full_date DATE NOT NULL,
    day INTEGER,
    month INTEGER,
    quarter INTEGER,
    year INTEGER
);
""")

conn.commit()

print("dim_date table created successfully!")

conn.close()
```

```python
#Fact NAV
```

```python
conn = sqlite3.connect("mutual_fund.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE fact_nav (
    nav_key INTEGER PRIMARY KEY AUTOINCREMENT,
    fund_key INTEGER NOT NULL,
    date_key INTEGER NOT NULL,
    nav_value REAL NOT NULL,

    FOREIGN KEY (fund_key)
        REFERENCES dim_fund(fund_key),

    FOREIGN KEY (date_key)
        REFERENCES dim_date(date_key)
);
""")

conn.commit()

print(" fact_nav table created successfully!")

conn.close()
```

```python
#Fact Transactions
```

```python
conn = sqlite3.connect("mutual_fund.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE fact_transactions (
    transaction_key INTEGER PRIMARY KEY AUTOINCREMENT,
    investor_id INTEGER NOT NULL,
    fund_key INTEGER NOT NULL,
    date_key INTEGER NOT NULL,

    transaction_type TEXT NOT NULL,
    amount REAL NOT NULL,
    units REAL,

    FOREIGN KEY (fund_key)
        REFERENCES dim_fund(fund_key),

    FOREIGN KEY (date_key)
        REFERENCES dim_date(date_key)
);
""")

conn.commit()

print(" fact_nav table created successfully!")

conn.close()
```

```python
# Fact Performance
```

```python
conn = sqlite3.connect("mutual_fund.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE fact_performance (
    performance_key INTEGER PRIMARY KEY AUTOINCREMENT,
    fund_key INTEGER NOT NULL,

    return_1yr_pct REAL,
    return_3yr_pct REAL,
    return_5yr_pct REAL,

    benchmark_3yr_pct REAL,
    alpha REAL,
    beta REAL,
    sharpe_ratio REAL,
    sortino_ratio REAL,
    std_dev_ann_pct REAL,
    max_drawdown_pct REAL,

    morningstar_rating INTEGER,
    expense_ratio_pct REAL,

    FOREIGN KEY (fund_key)
        REFERENCES dim_fund(fund_key)
);
""")

conn.commit()

print(" fact_nav table created successfully!")

conn.close()
```

```python
#Fact AUM
```

```python
conn = sqlite3.connect("mutual_fund.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE fact_aum (
    aum_key INTEGER PRIMARY KEY AUTOINCREMENT,
    fund_key INTEGER NOT NULL,
    date_key INTEGER NOT NULL,

    aum_crore REAL NOT NULL,

    FOREIGN KEY (fund_key)
        REFERENCES dim_fund(fund_key),

    FOREIGN KEY (date_key)
        REFERENCES dim_date(date_key)
);
""")

conn.commit()

print(" fact_nav table created successfully!")

conn.close()
```
