```python
#Standardise transaction_type values (SIP/Lumpsum/Redemption)
#.str.strip(), .str.title(), mapping variants, validation with isin()
```

```python
#2. Standardize transaction_type
```

```python
'''Common values should be:

SIP
Lumpsum
Redemption

Clean text first:
'''
```

```python
df['transaction_type'] = (
    df['transaction_type']
    .astype(str)
    .str.strip()
    .str.title()
)
```

```python
#Map variants:
```

```python
transaction_map = {
    'Sip': 'SIP',
    'S.I.P': 'SIP',
    'Systematic Investment Plan': 'SIP',

    'Lump Sum': 'Lumpsum',
    'Lumpsum': 'Lumpsum',

    'Redeem': 'Redemption',
    'Redemption': 'Redemption'
}

df['transaction_type'] = (
    df['transaction_type']
    .replace(transaction_map)
)
```

```python
#Validate:
```

```python
valid_types = ['SIP', 'Lumpsum', 'Redemption']

invalid_txn = df[
    ~df['transaction_type'].isin(valid_types)
]

print("Invalid transaction types:", len(invalid_txn))
print(invalid_txn['transaction_type'].unique())
```

```python
#Validate amount > 0
#df = df[df['amount_inr'] > 0]
```

```python
#3. Validate Amount > 0

#Check invalid values:
```

```python
df = df[df['amount_inr'] > 0]
```

```python
#Fix date formats
#pd.to_datetime(..., errors='coerce') 
#before date dtype is object and the convert in datetime64
```

```python
# 4. Fix Date Formats

# Convert to datetime:
```

```python
df['transaction_date'] = pd.to_datetime(
    df['transaction_date'],
    errors='coerce'
)
```

```python
df = df.dropna(subset=['transaction_date'])
```

```python
import matplotlib.pyplot as plt

plt.hist(df['amount_inr'], bins=30)

plt.title('Investment Amount Distribution')
plt.xlabel('Amount (INR)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Transactions Over Time
```

```python
monthly_txn = (
    df.set_index('transaction_date')
      .resample('M')
      .size()
)

plt.figure(figsize=(10,5))
plt.plot(monthly_txn.index, monthly_txn.values)

plt.title('Transactions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.tight_layout()
plt.show()
```
